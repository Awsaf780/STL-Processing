In this Notebook, STL files are converted to High-Resolution PNG Images using "StlToVoxel" library. The images are rescaled when sliced, so all STL files taken as Input will result in same dimensions. These combination of images are, after that, augmented.

For this example, 3D object is sliced into 200 images, each with resolution 256x256. (These variables are changable)

# Make Imports

```python
import stltovoxel as sv
from tqdm.notebook import tqdm
from PIL import Image
from numpy import asarray
import numpy as np
import os
from matplotlib import pyplot as plt
import tensorflow as tf
```

# Define Variables

```python
stl_dir = 'stl'
png_dir = 'outputs'

resolution = 200
resolution -= 3

resize_val = (256, 256)

dataset = []

stl_files = os.listdir(stl_dir)
stl_files
```

# Function to convert STL file to PNG Chunks

```python
def stl_to_png(filename, resolution=100):
    file = filename.replace('.stl', '')
    folder = '{}/{}'.format(png_dir, file)
    
    if not os.path.exists(folder):
        os.makedirs(folder)
        
    sv.convert_file('{}/{}'.format(stl_dir, filename), '{}/{}.png'.format(folder, file), resolution=resolution)
    return file
```

# Function to read PNG images per file as ND-Array

```python
def get_tensor(name, resize_val):
    path_prefix = '{}/{}/{}_'.format(png_dir, name, name)
    
    data_list = []
    global resolution
    for i in range(resolution + 3):
        filename = "{}{}.png".format(path_prefix, str(i).zfill(3))
        
        data = np.array(Image.open(filename).resize(resize_val)).astype('int')
        data_list.append(data)

    return np.array(data_list)
```

# Iterate all STL files and Prepare Dataset

```python
for file in tqdm(stl_files):
    try:
        name = stl_to_png(file, resolution)
        tensor = get_tensor(name, resize_val)
            
        dataset.append(tensor)
    except Exception as e:
        print(file + str(e))
```

```python
def plot_slices(num_rows, num_columns, width, height, data):
    
    """Plot a montage of 40 Teeth slices"""
#     data = np.rot90(np.array(data))
#     data = np.transpose(data)
    data = np.reshape(data, (num_rows, num_columns, width, height))
    
    rows_data, columns_data = data.shape[0], data.shape[1]
    heights = [slc[0].shape[0] for slc in data]
    widths = [slc.shape[1] for slc in data[0]]
    
    fig_width = 12.0
    fig_height = fig_width * sum(heights) / sum(widths)
    
    f, axarr = plt.subplots(
        rows_data,
        columns_data,
        figsize=(fig_width, fig_height),
        gridspec_kw={"height_ratios": heights},
    )
    
    for i in range(rows_data):
        for j in range(columns_data):
            axarr[i, j].imshow(data[i][j], cmap="gray")
            axarr[i, j].axis("off")
            
    plt.subplots_adjust(wspace=0, hspace=0, left=0, right=1, bottom=0, top=1)
    plt.show()
```

#### Plot Slices of Original Instance

```python
plot_slices(20, 10, 256, 256, instance[:200, :, :])
```

# Function to Rotate 3D array at random angles

```python
import random
from scipy import ndimage


@tf.function
def rotate(instance):
    """Rotate the instance by a few degrees"""

    def scipy_rotate(instance):
        # define some rotation angles
        angles = [-20, -10, -5, 5, 10, 20]
        # pick angles at random
        angle = random.choice(angles)
        # rotate instance
        instance = ndimage.rotate(instance, angle, reshape=False)
        instance[instance < 0] = 0
        instance[instance > 1] = 1
        return instance

    augmented_instance = tf.numpy_function(scipy_rotate, [instance], tf.int32)
    return augmented_instance
```

### Plot 2 sample augmentations

```python
augment1 = rotate(instance)
plot_slices(20, 10, 256, 256, augment1[:200, :, :])
```

```python
augment2 = rotate(instance)
plot_slices(20, 10, 256, 256, augment2[:200, :, :])
```
